```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

When I use TensorFlow, I define my layers using Sequential. Inside the Sequential, I specify what each layer looks like.
In a scenario such as this one, the computer has no idea what the relationship between X and Y is. So it will make a guess. Say for example it guesses that Y=10X + 10. It then needs to measure how good or how bad that guess is. That's the job of loos funciton. Source: AI and Machine Learning for Coders,Laurence Moroney. Copyright 2021 Laurence Moroney, 978-1-492-07819-7


```python
xs = np.array([-1.0,0.0,1.0,2.0,3.0,4.0], dtype=float)
ys = np.array([-3.0,-1.0,1.0,3.0,5.0,7.0], dtype=float)
model.fit(xs, ys, epochs=500)
```

```python
print(model.predict(np.array([10.0])))
print("Here is what I learned: {}".format(model.get_weights()))
```
